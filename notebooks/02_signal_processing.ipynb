```python
import pandas as pd
import os
from pathlib import Path

# Veri yolu
data_path = "d:/ecg/data/raw/ptbxl/"

# Meta veriyi oku
df = pd.read_csv(data_path + "ptbxl_database.csv")

# Dosya yollarını kontrol et
print("İlk 5 kayıt için dosya yolları:")
print(df['filename_hr'].head())

# Dizin yapısını kontrol et
print("\nDizin içeriği:")
for root, dirs, files in os.walk(data_path):
    level = root.replace(data_path, '').count(os.sep)
    indent = ' ' * 4 * level
    print(f"{indent}{os.path.basename(root)}/")
    if level < 2:  # Çok fazla dosya göstermeyelim
        for f in files:
            print(f"{indent}    {f}")
```
